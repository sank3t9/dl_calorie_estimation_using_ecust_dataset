```python
import os
import shutil

dataset_dir = "ecust_dataset"
splits = ["train", "val", "test"]

for split in splits:
    split_file = os.path.join(dataset_dir, "ImageSets", "Main", f"{split}.txt")
    image_output_dir = os.path.join(dataset_dir, "images", split)
    os.makedirs(image_output_dir, exist_ok=True)

    with open(split_file, "r") as f:
        lines = f.read().splitlines()

    for image_id in lines:
        image_file = os.path.join(dataset_dir, "JPEGImages", f"{image_id}.jpg")
        if os.path.exists(image_file):
            shutil.copy(image_file, image_output_dir)
```

```python
from ultralytics import YOLO

yolo_model = YOLO('yolov8s.pt')  

results = yolo_model.train(
    data='C:/Users/Sanket/Desktop/DL Final Project/ecust.yaml',
    epochs=30,
    patience=5,          
    imgsz=640,
    batch=16,
    save=True,
    save_period=1,
    name='ecust_yolov8_run',
    verbose=True
)
```

```python
from ultralytics import YOLO
model = YOLO("runs/detect/ecust_yolov8_run/weights/best.pt")
```

```python
metrics = model.val(split='test')
```

```python
print("Mean Average Precision (mAP) Metrics:\n")
print(f"mAP@0.5       : {metrics.box.map:.4f}")
print(f"mAP@0.5 (50)  : {metrics.box.map50:.4f}")
print(f"mAP@0.75 (75) : {metrics.box.map75:.4f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

results_path = "runs/detect/ecust_yolov8_run/results.csv"
df = pd.read_csv(results_path)

plt.figure(figsize=(12, 6))
plt.plot(df['metrics/mAP50(B)'], label='mAP@0.5')
plt.plot(df['metrics/precision(B)'], label='Precision')
plt.plot(df['metrics/recall(B)'], label='Recall')
plt.title("YOLOv8 Training Performance")
plt.xlabel("Epoch")
plt.ylabel("Score")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
names_dict = metrics.names
class_names = list(names_dict.values())

precision = metrics.box.p.tolist()
recall = metrics.box.r.tolist()
map_50 = metrics.box.map.tolist()
map_50_95 = metrics.box.map50.tolist()

df_cls = pd.DataFrame({
    "Class": class_names,
    "Precision": precision,
    "Recall": recall,
    "mAP@0.5": map_50,
    "mAP@0.5:0.95": map_50_95
})

df_cls = df_cls.round(3)
display(df_cls)
```

```python
df_cls.plot(x="Class", y=["Precision", "Recall"], kind="bar", figsize=(12, 6))
plt.title("Class-wise Precision and Recall")
plt.grid()
plt.show()
```

```python
precision = metrics.box.p.mean().item()
recall = metrics.box.r.mean().item()
f1_score = 2 * precision * recall / (precision + recall + 1e-6)

print(f"Overall Precision: {precision:.4f}")
print(f"Overall Recall   : {recall:.4f}")
print(f"Overall F1 Score : {f1_score:.4f}")
```
